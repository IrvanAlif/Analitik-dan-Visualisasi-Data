## **outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Tidak ada nilai yang berada di luar batas IQR (Interquartile Range). Dengan kata lain, semua data dalam kolom numerik berada dalam rentang yang dianggap normal menurut metode IQR.

```python
# Menyimpan daftar kolom awal sebelum modifikasi
original_columns = df.columns.tolist()

# Fungsi untuk menentukan kategori pengguna berdasarkan Daily_Screen_Time_Hours
def categorize_user(screen_time_hours):
    if screen_time_hours < 4:
        return "Light User"
    elif 4 <= screen_time_hours <= 8:
        return "Moderate User"
    else:
        return "Heavy User"

# Menambahkan kolom 'User_Category' tanpa mengubah struktur dataset lainnya
df["User_Category"] = df["Daily_Screen_Time_Hours"].apply(categorize_user)

# Memastikan format tetap 1000 rows × 10 columns (hanya mengganti kolom terakhir dengan User_Category)
df = df[original_columns + ["User_Category"]]  # menambahn 'User_Category'

# Mengecek bentuk akhir dataset
df.shape
```

```python

import numpy as np

# Contoh DataFrame dengan 1000 baris dan 11 kolom
data = {
    "User_ID": range(1, 1001),
    "Daily_Screen_Time_Hours": np.random.randint(1, 12, 1000),
    "User_Category": np.random.choice(["Light User", "Moderate User", "Heavy User"], 1000),
    "Age": np.random.randint(18, 60, 1000),
    "Gender": np.random.choice(["Male", "Female"], 1000),
    "Location": np.random.choice(["City", "Suburb", "Rural"], 1000),
    "App_Usage_Time": np.random.randint(0, 10, 1000),
    "Browsing_Time": np.random.randint(0, 10, 1000),
    "Gaming_Time": np.random.randint(0, 10, 1000),
    "Streaming_Time": np.random.randint(0, 10, 1000),
    "Online_Shopping_Time": np.random.randint(0, 10, 1000),
}

df = pd.DataFrame(data)

# **1. Sampling** (mengambil subset dari data)
df_sample = df.sample(frac=0.5, random_state=42)  # Ambil 50% data secara acak

# **2. Dimensionality Reduction** (hapus fitur yang kurang penting)
df_reduced = df.drop(columns=["Browsing_Time", "Gaming_Time"])  # Hapus 2 kolom

# **3. Data Aggregation** (buat kolom total screen time)
df["Total_Screen_Time"] = (
    df["App_Usage_Time"] + df["Streaming_Time"] + df["Online_Shopping_Time"]
)

# **4. Data Compression** (konversi tipe data untuk hemat memori)
df["User_ID"] = df["User_ID"].astype("int16")
df["Age"] = df["Age"].astype("int8")

# Cek ukuran dataset setelah reduksi
print("Ukuran dataset awal:", df.memory_usage(deep=True).sum(), "bytes")
print("Ukuran dataset setelah reduksi:", df_reduced.memory_usage(deep=True).sum(), "bytes")

```
